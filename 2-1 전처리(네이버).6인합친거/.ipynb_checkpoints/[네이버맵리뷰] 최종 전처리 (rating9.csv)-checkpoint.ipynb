```python
import pandas as pd
import numpy as np
```

```python
# csv 불러오기

df_nam = pd.read_csv('../data/final/nam_rating9.csv', encoding='utf-8-sig', index_col=0)
df_jun = pd.read_csv('../data/final/jun_rating9.csv', encoding='utf-8-sig', index_col=0)
df_mun = pd.read_csv('../data/final/mun_rating9.csv', encoding='utf-8-sig', index_col=0)

# 제주통합음식점 csv 불러오기

jeju = pd.read_csv('./제주특별자치도음식점목록(통합).csv', encoding='utf-8-sig', index_col=0)
```

# 전처리 내용
* 데이터프레임 컬럼 순서 변경
    - 상호명 / 도로명주소 / 지역분류 / 평점 / 작성자ID


* 업종소분류 컬럼 추가
    - 제주특별자치도음식점목록(통합).csv와 inner join하여 각 식당에 맞는 업종으로 추가
    - 게스트하우스 삭제

* 기타 삭제

```python
df_mun = df_mun[df_mun.상호명 != '느영나영']
df_mun = df_mun[df_mun.상호명 != '돈내코힐리조트']
df_mun = df_mun[df_mun.상호명 != '김군빌리지']
df_mun = df_mun[df_mun.상호명 != '사계리2033']
df_mun = df_mun[df_mun.상호명 != '유탑유블레스호텔제주']
df_mun = df_mun[df_mun.상호명 != '월정리펜션w728']
df_mun = df_mun[df_mun.상호명 != 'Hotel낙천주의자들']
df_mun = df_mun[df_mun.상호명 != '(주)김녕요트투어']
df_mun = df_mun[df_mun.상호명 != '더삼화스크린골프']
```

```python
df_jun.columns = ['작성자ID', '상호명', '평점']
df_jun = df_jun[['상호명', '평점', '작성자ID']]

df_nam.columns = ['작성자ID', '상호명', '평점']
df_nam = df_nam[['상호명', '평점', '작성자ID']]
```

```python
df = pd.concat([df_mun, df_jun, df_nam])
df
```

```python
df.reset_index(inplace=True)
df
```

```python
del df['index']
```

```python
# 도로명주소를 df로 추출

address = pd.DataFrame(jeju['도로명전체주소'])
address
```

```python
upjong
```

```python
address
```

```python
address
```

```python
# address 데이터프레임 전처리

# address.reset_index(inplace=True)
address.columns = ['상호명', '도로명주소']
address
```

```python
# upjong 데이터프레임 전처리

# upjong.reset_index(inplace=True)
upjong.columns = ['상호명', '업종소분류']
upjong
```

```python
# df에 도로명주소(address)와 업종소분류(upjong) 데이터프레임 inner join으로 concat

mer = pd.merge(df, address)
mer
```

```python
df = pd.merge(mer, upjong)
df
```

```python
# 상호명 / 도로명주소 / 지역분류 / 평점 / 작성자ID
df = df[['상호명', '업종소분류', '도로명주소', '평점', '작성자ID']]
df
```

```python
df.columns =['name', 'upjong', 'address', 'rating', 'id']
```
