```python
import googlemaps
import pandas as pd

my_key = "AIzaSyDa9AwBq_wIY4lraiCjJVf7CWgbHYyYyy0"
maps = googlemaps.Client(key=my_key)  # my key값 입력
lat = []  #위도
lng = []  #경도

# 위치를 찾을 장소나 주소를 넣어준다.
places = data2["address"]

i=0
for place in places:   
    i = i + 1
    try:
        print("%d번 인덱스에서 %s의 위치를 찾고있습니다"%(i, place))
        geo_location = maps.geocode(place)[0].get('geometry')
        lat.append(geo_location['location']['lat'])
        lng.append(geo_location['location']['lng'])
        

    except:
        lat.append('')
        lng.append('')
        print("%d번 인덱스 위치를 찾는데 실패했습니다."%(i))


# 데이터프레임만들어 출력하기
data3 = pd.DataFrame({'위도':lat, '경도':lng}, index=places)
print(data3)
```

```python
inner=pd.merge(data2,data3,on='address', how='inner')
```
