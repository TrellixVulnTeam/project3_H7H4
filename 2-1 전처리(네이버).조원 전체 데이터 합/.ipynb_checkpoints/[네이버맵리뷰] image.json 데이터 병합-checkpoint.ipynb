# 전체 image.json 파일 병합

```python
import pandas as pd
```

```python
one = pd.read_csv('./navermap_image/image.csv', encoding='utf-8-sig', index_col=0)
two = pd.read_csv('./navermap_image/음식점이미지url_정문_종민_남경.csv', encoding='utf-8-sig', index_col=0)
thr = pd.read_csv('./navermap_image/음식점이미지url_준영_용호_남경.csv', encoding='utf-8-sig', index_col=0)
```

```python
one.columns = ['rest_name', 'url']
one
```

```python
two
```

```python
thr
```

```python
df = pd.concat([one, two, thr])
df
```

```python
df.drop_duplicates(['rest_name', 'url'])
df
```

```python
df.reset_index(inplace=True)
df
```

```python
del df['index']
```

```python
df.to_csv('네이버맵_리뷰이미지_url합본.csv', mode='w')
```
