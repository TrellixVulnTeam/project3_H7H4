```python
import json
import pandas as pd
import numpy as np
import os
```

```python
basic = pd.read_csv('음식점이미지url_정문.csv', encoding='utf-8-sig', index_col=0)
basic
```

```python
# 크롤링한 image.json 파일 불러오기

with open("../data/nam_image.json", "r", encoding="utf-8") as json_file:
    json_sample = json.load(json_file)
    
# with open("../driver & crawling/느시/image.json", "r", encoding="utf-8") as json_file:
#     json_sample2 = json.load(json_file)
    
with open("../driver & crawling/느시/image_2nd.json", "r", encoding="utf-8") as json_file:
    json_sample3 = json.load(json_file)
```

```python
df = pd.json_normalize(json_sample)
# df2 = pd.json_normalize(json_sample2)
df3 = pd.json_normalize(json_sample3)
```

```python
df = df.T
# df2 = df2.T
df3 = df3.T
```

```python
df.reset_index(inplace=True)
# df2.reset_index(inplace=True)
df3.reset_index(inplace=True)
```

```python
df = pd.concat([df, df3])
df
```

```python
df.columns = ['name', 'url']
```

```python
df['url'].replace('', np.nan, inplace=True)
df = df.dropna(subset = ['url'])
```

```python
df['rest_name'] = df.name.str.split('.').str[0]
df
```

```python
df.reset_index(inplace=True)
```

```python
# del df['name']
df = df[['rest_name', 'url']]
df
```

```python
final = pd.concat([df, basic])
final
```

```python
final.reset_index(inplace=True)
final
```

```python
del final['index']
```

```python
final.to_csv('음식점이미지url_정문종민남경.csv', mode='w')
```
