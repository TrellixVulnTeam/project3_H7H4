```python
rest = pd.read_csv('./네이버맵음식점리뷰_평점_정문.csv', encoding='utf-8-sig', index_col=0)
rest
```

```python
rest.columns = ['작성자ID', '상호명', '평점']
rest = rest[['상호명', '평점', '작성자ID']]
rest
```

```python
# 느영나영게스트하우스 삭제
rest = rest[rest.상호명 != '느영나영']
```

```python
# 돈내코힐리조트, 감군빌리지, 사계리2033 삭제

rest = rest[rest.상호명 != '돈내코힐리조트']
rest = rest[rest.상호명 != '감군빌리지']
rest = rest[rest.상호명 != '사계리2033']
rest
```

```python
# (주)김녕요트투어 삭제
rest = rest[rest.상호명 != '(주)김녕요트투어']
rest
```

```python
rest.reset_index(inplace=True)
del rest['index']
```

```python
rest
```

```python
# 남경님 크롤링한 csv 데이터 불러오기

one = pd.read_csv('../data/nam_rating9.csv', encoding='utf-8-sig', index_col=0)
two = pd.read_csv('../driver & crawling/느시/rating9.csv', encoding='utf-8-sig', index_col=0)
thr = pd.read_csv('../driver & crawling/느시/rating9_2nd.csv', encoding='utf-8-sig', index_col=0)
```

```python
del one['Timestamp']
del two['Timestamp']
del thr['Timestamp']
```

```python
df = pd.concat([one, two, thr])
df
```

```python
# 리조트 및 기타 데이터 삭제
df = df[df.ItemID != '유탑유블레스호텔제주']
df = df[df.ItemID != '월정리펜션w728']
df = df[df.ItemID != 'Hotel낙천주의자들']
df = df[df.ItemID != '더삼화스크린골프']
df
```

```python
df.reset_index(inplace=True)
```

```python
df
```

```python
del df['index']
```

```python
df.columns = ['작성자ID', '상호명', '평점']
df = df[['상호명', '평점', '작성자ID']]
df
```

```python
# 정문 + 종민 + 남경 평점 데이터 합치기
df = pd.concat([df, rest])
df
```

```python
df.reset_index(inplace=True)
```

```python
del df['index']
```

```python
df.to_csv("네이버맵음식점리뷰_평점_정문_종민_남경.csv", mode='w')
```
