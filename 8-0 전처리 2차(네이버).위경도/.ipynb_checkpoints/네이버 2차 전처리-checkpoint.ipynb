```python
import pandas as pd
import numpy as np
```

```python
naver = pd.read_csv('./211020_네이버맵음식점평점.csv', encoding='utf-8-sig', index_col=0)
```

```python
naver
```

```python
naver['평점_건수']    = naver.groupby(['name'])['rating'].transform('count')
naver['평점_합계']    = naver.groupby(['name'])['rating'].transform('sum')
naver['평점_평균']    = naver.groupby(['name'])['rating'].transform('mean')
naver['평점_표준편차'] = naver.groupby(['name'])['rating'].transform('std')

```

```python
naver
```

```python
naver=naver.drop_duplicates(['name'])
```

```python
naver
```

```python
naver=naver[['name','upjong','address','category','rating','평점_건수','평점_합계','평점_평균','평점_표준편차']]
```

```python
naver
```

```python
result = pd.read_csv('./result1.csv', encoding='utf-8-sig', index_col=0)
```

```python
result 
```

```python
total=pd.merge(naver,result ,on='name', how='outer')
```

```python
total.to_csv('total.csv',mode="w",encoding='utf-8')
```
